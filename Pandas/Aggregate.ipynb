İleri Yoplulaştırma İşlemleri(Aggregate,filter,transform,apply)

```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame({'gruplar': ['A','B','C','A','B','C'],
                  'değişken1': [10,23,33,22,11,99],
                  'değişken2': [100,253,333,262,111,969]}, columns = 
                 ['gruplar', 'değişken1', 'değişken2'])
df
```

```python
#aggregate
```

```python
df.groupby("gruplar").mean()
```

```python
df.groupby("gruplar").aggregate(["min",np.median,max])
```

```python
df.groupby("gruplar").aggregate([min,np.median,max])
```

```python
df.groupby("gruplar").aggregate({"değişken1": "min", "değişken2": "max"})
```
